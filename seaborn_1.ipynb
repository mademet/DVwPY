```python
# Basic Library
import numpy as np
import pandas as pd
# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings;
warnings.filterwarnings('ignore')
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
fig, ax = plt.subplots()
ax.bar(langs, students)
for i in range(len(langs)):
    plt.text(langs[i], students[i], str(students[i]), ha="center",va="bottom")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x='total_bill',y='tip',data=tips)
ax.set(xlabel='Total Bill', ylabel='Tip', title='Some title')
#plt.legend(labels=['legend1'])
plt.show()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex", s=200);
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex", s=200, style="sex");
```

```python
sns.scatterplot(x='total_bill',y='tip',size="sex", data=tips);
```

```python
sns.scatterplot(x='total_bill',y='tip',s=200, data=tips, alpha=1);
#sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill',y='tip', data=tips,style='sex',hue='sex',s=100);
```

```python
sns.rugplot(x='total_bill', data=tips, height=0.5)
# height : number
```

```python
sns.rugplot(x='total_bill', data=tips, height=0.5);
# height : number
#     Proportion of axes extent covered by each rug element.
```

```python
sns.displot(x='total_bill', data=tips);
```

```python
sns.displot(x='total_bill', data=tips, kde=True);
```

```python
sns.displot(x='total_bill', data=tips, kde=True, bins=30);
```

```python
sns.set(style='darkgrid')
sns.histplot(data = tips, x = 'total_bill',bins=20);
```

```python
sns.kdeplot(x = "total_bill", data = tips, bw_adjust = 0.2, shade = True, color = "red");
```
