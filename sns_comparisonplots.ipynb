```python
# Basic Library
import numpy as np
import pandas as pd
# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings;
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.heatmap(tips.corr());
```

```python
sns.heatmap(tips.corr(),annot=True);
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.catplot(x = "pclass", y = "survived", data=titanic, kind="bar", hue="sex",
            estimator=np.sum, col = "alone", row = "embarked");
```

```python
tips
```

```python
sns.catplot(x = "day", y = "total_bill", data=tips, kind="box", hue="smoker");
```

```python
sns.catplot(x = "day", y = "total_bill", data=tips, kind="swarm", hue="smoker");
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex");
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde");
```

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips, hue="sex", palette="YlOrBr", corner=True);
```

```python
g = sns.PairGrid(tips);
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot, color="red")
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

```python
a = sns.pairplot(tips)
a.map_lower(sns.kdeplot, color='purple')
a.map_upper(sns.kdeplot, color='g');
```
