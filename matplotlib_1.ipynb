```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
plt.show()
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, z)
ax.plot(x, t)
ax.set_xlabel("X")
ax.set_ylabel("Y - Z")
ax.set_title("TITLE")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x, z)
plt.plot(x,t)
plt.xlabel("x")
plt.ylabel("y,z,t")
plt.title("x ' e göre y,z,t");
plt.show()
```

```python
fig = plt.figure()      # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0, 2, 2])
ax2 = fig.add_axes([2.1, 1.6, 0.4, 0.4])
ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")
ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

```python
# functional
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(age, salary, "r")
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b")
plt.subplot(2,2,3)
plt.plot(age, salary, "g" )
plt.subplot(2,2,4)
plt.plot(age, salary_2, "yellow")
plt.tight_layout();
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize= (16,4))
ax[0][1].plot(age, salary, "r")
ax[1][2].plot(age, salary_2, "b")
plt.show()
```

```python
plt.plot(age, salary, "r", ls=":", lw=3, marker= "o", markersize=10);  # lw or linewidth, ls or linestyle
```

```python
fig, ax = plt.subplots(figsize=(12,6))
fig.suptitle("To change the line width, we can use the `linewidth` or `lw` keyword argument",fontsize=18)
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10)
plt.show()
```

```python
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# Use marker for string code
# Use markersize or ms for size
ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 1.2)
plt.subplot(2,2,1)
plt.plot(age, salary, "r", ls="-", marker="+" ,markersize=30)
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b",ls="-.", marker="o" ,markersize=30)
plt.subplot(2,2,3)
plt.plot(age, salary, "g",ls=":", marker="x" ,markersize=30)
plt.subplot(2,2,4)
plt.plot(age, salary_2, "y",ls="--" ,marker="*" ,markersize=30)
plt.tight_layout();
plt.show()
```

```python
# Plot Range with set_ylim & set_xlim
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([3,5])
ax[1].set_ylim([1,650])
ax[0].axvline(x=5, ls="--")
ax[0].axvline(x=3, ls="--")
plt.show()
```

```python
plt.figure()

a1 = plt.subplot2grid((5,5),(0,0),colspan=3, rowspan=2)
a2 = plt.subplot2grid((5,5),(0,3),rowspan=5)
a3 = plt.subplot2grid((5,5),(2,0),rowspan=3, colspan=3)
a4 = plt.subplot2grid((5,5),(0,4),rowspan=5)

x = np.arange(1,10)

a1.plot(x, np.exp(x))
a1.set_title('exp')

a2.plot(x, x*x)
a2.set_title('square')

a3.plot(x,np.log(x))
a3.set_title('log')

a4.plot(x,np.sqrt(x))
a4.set_title('sqrt')

plt.tight_layout()
plt.show()
```
